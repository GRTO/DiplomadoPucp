# Pretratamiento de la data y visualizacion de indicadores

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))


places = df[df['Site Percentile'].str.contains('No. of sites')].drop(['Site Percentile'], axis=1)
places
```

```
columns = list(places.columns[1:])

metals = list(places.Metals)

plt.figure(figsize=(20,6))
for metal in metals:
  metalsValues = places[places.Metals == metal].values.tolist()[0][1:]
  plt.plot(metalsValues, marker='*', linewidth=2, alpha=0.9)
  plt.xticks(np.arange(len(columns)), columns, rotation=45)

plt.ylabel('N° Cities where metals were found')
plt.legend(metals)
plt.show()
```

**2. Indicador:** Esta grafica muestra la distribución de metales con **90th percentile**

```
percentilesMetals = df[~df['Site Percentile'].str.contains('No. of sites')]
percentilesMetals = percentilesMetals[percentilesMetals['Site Percentile'] == '90th percentile']

plt.figure(figsize=(20,6))
for metal in metals:
  percentilesMetalsValues = percentilesMetals[percentilesMetals.Metals == metal].values.tolist()[0][2:]
  plt.plot(percentilesMetalsValues, marker='*', linewidth=2, alpha=0.9)
  plt.xticks(np.arange(len(columns)), columns, rotation=45)
  
plt.ylabel('90th percentile Metals')
plt.legend(metals)
plt.show()
```

**3. Indicador:**  Indicador: Esta grafica muestra la distribución de metales con **50th percentile**

```
percentilesMetalsF = df[~df['Site Percentile'].str.contains('No. of sites')]
percentilesMetalsF = percentilesMetalsF[percentilesMetalsF['Site Percentile'] == '50th percentile']

plt.figure(figsize=(20,6))
for metal in metals:
  percentilesMetalsValuesF = percentilesMetalsF[percentilesMetalsF.Metals == metal].values.tolist()[0][2:]
  plt.plot(percentilesMetalsValuesF, marker='*', linewidth=2, alpha=0.9)
  plt.xticks(np.arange(len(columns)), columns, rotation=45)
  
plt.ylabel('50th percentil Metals')
plt.legend(metals)
plt.show()
```
